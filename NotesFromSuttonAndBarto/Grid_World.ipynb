```python
import numpy as np
```

```python
def p(iy,ix,a):
    r=0
    #iy,ix=i//5, i%5
    
    if a==4:
        return iy,ix,0
    
    if ix==1 and iy==0:
        r=10
        ixp=ix
        iyp=4
    elif ix==3 and iy==0:
        r=5
        ixp=ix
        iyp=2
    elif a==0:
        ixp=ix
        iyp=iy-1
    elif a==1:
        ixp=ix-1
        iyp=iy
    elif a==2:
        ixp=ix
        iyp=iy+1
    elif a==3:
        ixp=ix+1
        iyp=iy

    if ixp<0:
        ixp=0
        r=-1
    elif ixp>4:
        ixp=4
        r=-1
    elif iyp<0:
        iyp=0
        r=-1
    elif iyp>4:
        iyp=4
        r=-1
    return iyp,ixp,r
        
    
```

```python
vs=np.zeros((5,5))
gamma=.9
for _ in range(100):
    for iy in range(5):
        for ix in range(5):
            v=vs[iy][ix]
            z=[]
            for a in range(5):
                fy,fx,r = p(iy,ix,a)
                z.append(r +  gamma*vs[fy][fx])
            vs[iy][ix]=np.max(z)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

ax = sns.heatmap(vs, annot=np.asarray(['{:.1f}'.format(x) for x in vs.flatten()]).reshape(vs.shape),fmt="",linewidth=0.5)
plt.show()
```
